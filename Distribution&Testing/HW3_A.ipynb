```python
from __future__ import print_function
__author__ = 'Alexis Soto-Colorado'

import pylab as pl
import numpy as np
import seaborn as sns
%pylab inline
```

```python
mysizes=np.random.random_integers(10,2000,100)
print (mysizes)
```

```python
dnormal=numpy.random.normal(loc=100,scale=1.0,size=100)
print(dnormal)
pl.hist(dnormal)
```

```python
bnominal=np.random.binomial(100,0.5,2000)
print (bnominal)
plt.hist(bnominal) 
```

```python
poison=np.random.poisson(lam=1.0, size=500)
print(poison)
plt.hist(poison)
```

```python
mymean = 100 
distributions = {}
distributions['chisq'] = np.random.chisquare(mymean, 100)
```

```python
pl.hist(distributions['chisq'])
```

```python
bt=np.random.gamma(2, 900,100)
print(bt)
plt.hist(bt)
#paleto=np.random.pareto(2, 900)
#print(paleto)
#plt.hist(paleto)
```

```python
sip=np.random.exponential(5, 900)
print(sip)
plt.hist(sip)
```

```python
graphmeans = []
distributions={'dnormal':{},'poison':{},'binomial':{},'chisq':{},'bt':{},'sip':{}}
for n in mysizes:
    distributions['dnormal'][n] = np.random.normal(mymean,1,n)
    distributions['poison'][n] = np.random.poisson(mymean,n)
    distributions['binomial'][n] = np.random.binomial(2*mymean,0.5,n)
    distributions['chisq'][n] = np.random.chisquare(mymean,n)
    distributions['bt'][n] = np.random.gamma(mymean,1,n)
    distributions['sip'][n] = np.random.exponential(mymean, n)
    
for n in mysizes:
    graphmeans.append(distributions['dnormal'][n].mean())
    graphmeans.append(distributions['poison'][n].mean())
    graphmeans.append(distributions['binomial'][n].mean())
    graphmeans.append(distributions['chisq'][n].mean())
    graphmeans.append(distributions['bt'][n].mean())
    graphmeans.append(distributions['sip'][n].mean())
```

```python
means = {}
means['dnormal'] = {}

fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions ['dnormal']: 
    axchisq_mu_n.plot(nn,distributions['dnormal'][nn].mean(),'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```
